Digital Humanities, Higher Education,

It has become widely accepted that students in the humanities need a grounding in digital techniques, both to stay abreast of developments in their own dosciplines and to ensure their employability after graduation. Thus far, few humanities courses in the low countries actually include any systematic teaching to arrive at such a grounding. In this article, based on our experiences at the Erasmus University Rotterdam, we identify obstacles to digital education of humanities majors and how we have overcome them.

## Introduction

```python
“(…) anything worth doing is worth doing badly.”
Jack Gilbert, Refusing Heaven (2005)
```

Since the turn of the century, digital historical scholarship has expanded significantly worldwide. Governments and research consortia across the globe, but especially in the Global North, have invested heavily in the establishment of dedicated research infrastructures for the humanities and social sciences, and for digital history in particular. Both the mass digitization of existing source corpora and the increased preservation of born-digital materials have created an ever-growing body of data for historians to analyse. And digital historians have gradually solidified digital history as a significant subfield, through the establishment of research centres such as the Roy Rosenzweig Center for History and New Media (RRCHNM) (US, 1994) and the Centre for Contemporary and Digital History (C2DH) (Luxembourg, 2016), the organization of dedicated conferences on digital history, and the establishment of academic journals such as the Journal of Digital History (2022).

As part of this effort, digital historians have formulated essential philosophical and methodological benchmarks for digital historical scholarship, while continuing to develop an ever-expanding set of digital historical research methodologies. Zaagsma <cite id="nc5ki"><a href="#zotero%7C1912747%2F2BCAFZVA">(2023)</a></cite> draws from insights in archival, library, and information science to provide an in-depth discussion of the ‘politics of digitization’, or how inequalities in the access to digital source corpora impact the practice of historical research worldwide. Hoekstra and Koolen<cite id="a13mr"><a href="#zotero%7C1912747%2FFQ58WDXV">(2019)</a></cite> discuss the importance of a ‘data scope’, i.e., an explicitly mapped out set of research question-driven data transformations, involving selection, modelling, linking, normalization an classification of data, when carrying out historical research with digital corpora. Both these discussions add to a wider range of methods in diverse areas such as natural language processing (NLP) (e.g., Garg et al. <cite id="9rez7"><a href="#zotero%7C1912747%2F4G3JGWR4">(2018)</a></cite>; , Guldi <cite id="32vzn"><a href="#zotero%7C1912747%2FXIISM5PU">(2019)</a></cite>; <cite id="mytru">Mendelsohn et al.<a href="#zotero%7C1912747%2FBHQIQXZD">(2020)</a></cite>; Wevers & Koolen<cite id="182rf"><a href="#zotero%7C1912747%2F4BTWWVXI">(2020)</a></cite>), visual analysis (e.g., <cite id="ezu7k"><a href="#zotero%7C1912747%2FH77XQIBG">(Smits, 2025)</a></cite>; <cite id="zh85u"><a href="#zotero%7C1912747%2FIYFSAWYQ">(Smits &#38; Wevers, 2023)</a></cite>), and computational web archive research (Brügger & Milligan, 2018; Verhoef<cite id="eeng2"><a href="#zotero%7C1912747%2FWQ6WITEZ">(2025))</a></cite>, to allow for what Milligan, Weingart and Graham<cite id="6tf0f"><a href="#zotero%7C1912747%2FUV7SY4N3">(2016)</a></cite> have identified as ‘macroscopic’ historical research.

These developments showcase the general dynamism and growth of digital history, and digital historical research methodologies are certainly beginning to gain wider traction (e.g. <cite id="0amxp"><a href="#zotero%7C1912747%2FBDZRILJQ">Huijnen, 2025)</a></cite>. But a key observation anno 2025 remains that the use of these methodologies within the historical profession as a whole continues to be limited. Many historians, while certainly making use of the various digital repositories that have been made available to them over the past two decades, still do so in a rather open-ended manner, while continuing to draw from older, often implicit and loosely demarcated, interpretative historical traditions. As a result, one can still meaningfully reference in 2025 what Zaagsma already wrote over a decade ago concerning digital historical scholarship, in only a slightly modified manner: “What we have then is a small vanguard of self-described digital historians, whereas the discipline as a whole is (still) struggling to come to terms with the brave new world” <cite id="qxhsr"><a href="#zotero%7C1912747%2FCP55TAX5">(Zaagsma, 2013, p.5)</a></cite>.

This limited adoption and impact of digital research methodologies in the history discipline can, at least in part, be attributed to how historians are currently still trained. In undergraduate history programs, digital history methods are rarely systematically taught, but rather relegated to isolated and optional elective courses or minor programs. And in graduate programs, courses on digital methods are either fully absent or integrated in a very limited number of specialized tracks on digital humanities, as teachings on these methods are considered to be part of computer science or other fields. The history programs offered at Dutch universities can serve as a direct illustration in this regard. Of all bachelor programs in history offered in the Netherlands, only one includes a mandatory course on the use of ‘big data’ in the core curriculum (Utrecht University, 2025), while most master programs do not offer any mandatory teaching in digital history. In the Netherlands, only one dedicated master program on digital humanities is currently in existence (University of Groningen, 2025). This overview is exemplary for the situation in other countries as well.

This apparent disinterest in digital history approaches in history programs stands in stark contrast to the extensive attention for education amongst digital humanists and digital historians themselves. As recently (again) highlighted by Croxall and Jakacki <cite id="g4gze"><a href="#zotero%7C1912747%2F3U4C4HPJ">(2023)</a></cite>, pedagogy and didactics have been a central part of the debate on digital humanities and digital history for a very long time. Scholars in these fields have been sharing their teaching experiences and developing instructional materials via blog posts and repositories, conference panels, handbooks and special journal issues, and more, over at least the past two decades. For digital history in particular, The Programming Historian, an open source outlet that offers access to various lessons on the use of digital research methods, has existed since 2008, while dedicated handbooks on the topic are increasingly available (see for example: Guiliano <cite id="d4g3h"><a href="#zotero%7C1912747%2FIBV7443M">(2022)</a></cite>. Apart from this, scholars specialized in the use of specific programming languages and methods have been developing dedicated materials, as illustrated by a rapidly growing number of handbooks on text mining methods.

In this article, we depart from the premise that digital historical research methods are a core component of a contemporary historian’s basic toolkit, and that a more widespread adoption of these methods is desirable. This implies that history educators need to find ways to systematically integrate their use in (under)graduate-level history programs. To contribute to this effort, we discuss a number of major lessons we have learned in this regard, as we have been teaching a digital history course in the (under)graduate history programs offered at Erasmus University Rotterdam in the Netherlands over the past five years.

In what follows, we first provide a general overview of the current set-up of our course and how it came into being. Next, we discuss the key pedagogic and didactic principles we have come to implement in our course based on our experiences of introducing digital history methods to (under)graduate students with no prior knowledge. We do so by first identifying broader problems educators quickly encounter when teaching digital history, and by then discussing what we have found to be productive remedies for them. Finally, we highlight what we consider to be remaining key challenges, and how we hope to address them going forward.

## General Overview of the Course

The foundations for our course on digital history at Erasmus University Rotterdam were established during the 2020-2023 Covid-19 pandemic. In the Fall of 2020, one of the authors of this article was asked to set up a temporary elective course on digital history in the Bachelor history program, to fill up a gap that had arisen due to the pandemic. At the time, the students enrolled in the program experienced major difficulties in pursuing an internship, then an obligatory component of the Bachelor program, as lockdowns had significantly reduced public life. In the academic year 2021-2022, the course was again shelved, as lockdown measures in the Netherlands were gradually phased out and universities, organizations in the cultural sector and most other institutions were again allowed to fully re-open in January 2022. However, during that same year, we were asked to contribute as teachers to a Winter School on digital history organized by the Huizinga Institute, the Dutch national research school for cultural historians (Huizinga Institute, 2021). Simultaneously, as the department management team had become appreciative of the instruction on digital history we had developed over the past years, the department decided to integrate a new version of the course on digital history in the Master history programs offered at Erasmus University Rotterdam, as an obligatory course for all Master students, in the academic year 2022-2023. Since then, both authors have been teaching the digital history course at the graduate level, aided by other colleagues at Erasmus University Rotterdam and beyond.

When we started teaching on digital history, we had varying levels of expertise and experience in the field, and in digital history education in particular. This especially applies to the use of programming languages such as R and Python for digital history. One of the authors had no prior expertise or experience in coding and developed instruction on digital history and coding through trial and error. This aligns with what McGinn and Coats describe as ‘born-pedagogical Digital Humanities (bpDH), or a DH-pedagogy that “arises from the space of curiosity, learning, and experimentation that is the classroom” <cite id="rl70b"><a href="#zotero%7C1912747%2F4HAFLDKW">(McGinn et al., 2023)</a></cite>). The other author had some prior expertise and experience in digital history, for example through the supervision of a PhD research project on the use of text mining-methods such as sentiment analysis to investigate emotions in political debates. However, also for the second author, digital history education was relatively new in comparison to research.

When first setting up the course on digital history in the Bachelor program, it only included a limited amount of hands-on coding. The course included sessions on writing and other forms of knowledge dissemination in the digital age, the use of digital research infrastructures such as the CLARIAH Media Suite (a research environment within the Dutch CLARIAH infrastructure dedicated to various audiovisual collections) and the use of the ATLAS.Ti software for qualitative data analysis. Apart from this, the course included two dedicated sessions on using the programming language R for text mining. During these two sessions, students learned about the RStudio interface, basic operations in R, how to load and clean textual data in R, and entry-level methods such as concordancing, word frequency analysis, and the use of lexical dispersion plots.

For the course in the Master programs, the set-up was significantly changed. A stronger emphasis was placed on the hands-on use of the programming language R, through the inclusion of dedicated sessions on digital research methods such as web scraping, sentiment analysis, topic modelling, and the use of word embedding models. This set-up was then adapted in subsequent years. For example, as part of an overarching effort to strengthen methodological instruction in all Master history programs at Erasmus University Rotterdam in 2024, two general sessions were added on methodological instruction, in particular on source criticism in the digital age and on operationalization. In addition, over the years, in light of the availability of other colleagues involved in the course, some of the method-specific sessions were changed, for example by replacing a session on web scraping by one on network analysis. An overview of the current set-up of the course can be found in Table 1.

Since we started teaching it, the Master course has consisted of a total of eight sessions of three times forty-five minutes. Initially, we adopted the following overall set-up for every session dedicated to a specific digital method. In the first part of the session, we introduced the method in general terms, by discussing the underlying principles, strengths and limitations of the method and other (epistemological, operational,…) matters. In the second part, we further expanded on this initial overview, by organizing a wider discussion with our students based on the course readings. Finally, in the third part, we demonstrated the discussed method through the use of the programming language R, by running a pre-prepared script live in class. Upon completion of the demonstration, we asked the students to use the scripts themselves in RStudio and further experiment with them by tailoring them to their interests, use different datasets, and more, while we provided coaching. Given that this original set-up repeatedly resulted in a lack of time for the students to experiment with the scripted material in class, we gradually reduced the amount of time dedicated to discussing the literature in class, to allow for more hands-on work.

In terms of assessment, we adopted a combination of two assignments, a set-up we still use today. For the first assignment, the students need to write a number of blog posts, to be shared with fellow students via the online learning environment of the course. In these blog posts, the students need to critically reflect on theoretical or methodological principles discussed during the preceding sessions of the course. As part of this assignment, the students are encouraged to adopt a ‘share your learning’-curve approach, as introduced by Luyendijk <cite id="w0s4h"><a href="#zotero%7C1912747%2FMBPXBTDM">(2011)</a></cite>. This approach entails that the students take their current level of knowledge about digital history and coding with R as a starting point, to subsequently track their insights and progress online. As a part of the assignment, the students also need to extensively comment at least twice per submission on each other’s blog posts, to encourage peer-to-peer learning.

For the second assignment, the students need to write a research paper in which they carry out a small-scale historical analysis by using at least one of the digital research methods that were discussed during the course. For the paper assignment, the students also need to develop a fitting personalized code base to support their analysis. Over the years, we have come to position the paper assignment as a ‘feasibility study’ first and foremost. The students are required to formulate a research idea that departs from a proper epistemological and operational understanding of the chosen digital research method. Next, they need to report on the strengths and limitations of their chosen approach and whether or not it proved to be feasible, within the constraints of the overall workload for the course. Beyond these formal criteria, the students are actively encouraged to experiment with the provided code and other course materials and develop their own research ideas, also in relation to the research they will carry out for a Master thesis, the main component of the Master programs.

## Paving the Road to Happiness: Didactic Principles for Digital History Education

By teaching the course for several years and adapting it based on student feedback and personal experience, we have come to learn several lessons on what to do, and what not to do, when teaching a course on digital history, in particular to graduate-level students with entry-level knowledge on the topic. In the following paragraph, we discuss a number of these lessons, also in light of the existing scholarship on pedagogies and didactics in digital humanities and digital history. We first identify key problems we encountered and aimed to address while teaching, and then discuss how we have tried to do so.

### White (R-)lines (just do it): on computational literacy, and serving as a ‘gateway drug’ for digital history

A key issue for an entry-level course on digital history today is that most people who enter the classroom have grown up in an environment where digital technology is not only often carefully curated through sandboxed and restrictive user interfaces, but also primarily designed for ease of use. This has diminished the ability of students to tailor digital technology to their own needs, which was in many ways a key characteristic of not only early computing, but also of many contemporary innovations in digital history. It has equally led students to adopt various use practices that are limited and even counterproductive when carrying out historical research, and that actively work against a better understanding of many of the digital historical research methodologies we want to teach.

For example, in a recent study on the use of the CLARIAH Media Suite, the Dutch digital research environment for audiovisual collections mentioned above, Sauer, Hagedoorn and Aasman (2024) demonstrate how students initially tend to struggle significantly when using the Media Suite as a curated archive, rather than as a data-rich social media platform. As students have come to adopt search habits that are mostly informed by Google and YouTube (and increasingly, generative AI), and expect fast results based on direct queries, they first need to ‘de-Google’ their search habits. They need to learn how to approach the Media Suite as a digital environment structured along more traditional archival principles, rather than algorithmic architectures that produce probability outcomes and integrate path dependencies, such as personalization.

When teaching digital history, we have made similar observations as  Sauer, Hagedoorn and Aasman (2024). Students have come to adopt a digital ‘habitus’ that needs fundamental correcting: they are in need of a better overall understanding of the digital environments they navigate and digital instruments they use, also in particular if they want to successfully carry out historical research. As a result, we decided to make computational understanding in itself an important part of the course taught in the Master. This means that we not only teach about the pitfalls of digitization and broader principles underpinning digital historical research methods. We equally decided to familiarize our students with coding in general terms, in particular through the use of the programming language R and its integrated development environment (IDE), RStudio.

Whether students who want to engage in digital humanities- and digital history scholarship should learn how to code, with R, Python, or another programming language, has been the topic of extended discussion among digital humanists and digital historians over the past two decades (see for example: Ramsay, 2013). We decided to do so because it helps students understand computational logic and develop a computational mode of thinking (e.g.,<cite id="5lhkl"><a href="#zotero%7C1912747%2FET79ZQ3E">(Wing, 2006)</a></cite> Wing, 2006). This contributes to the development of the computational literacy necessary to understand the architectures and operationalizations underpinning the infrastructures and methodologies discussed in class (e.g., diSessa, 2000), as well as, for example, the ‘politics of digitization’ highlighted above (Zaagsma, 2023). This argument was complemented by labour market considerations, in that integrating coding instruction offers students competences that improve their employability. This argument plays a more direct role in the positioning of our course within the Master history programs offered at Erasmus University Rotterdam as a whole (for the full paragraph, see also:<cite id="yt1ah"><a href="#zotero%7C1912747%2FQ6ZJTLWS">(Blikstein &#38; Moghadam, 2019)</a></cite>.

The choice to familiarize students with coding raised the question how we would do so in practice. Here, we of course bumped into the limitation that our course is not a fully-fledged programming course and could never be one. Proper in-depth coding would require broader curriculum development, which falls outside the scope of both our course and the Master history programs offered at Erasmus University Rotterdam in general. However, we did decide to make coding a central pillar of our instruction, in that we not only introduce basic coding concepts and demonstrate coding live in class, but also ask students to develop their own code base throughout the course. This particular combination allows us to achieve two goals. On the one hand, it offers students the starting points they need to sharpen their computational literacy skills. On the other hand, asking them to develop their own code base compels them to further delve into the learning materials themselves and make them their own. This sets them up for what educational scholars such as <cite id="aetxp"><a href="#zotero%7C1912747%2FHY79A3XX">(Boud, 2012)</a></cite>, <cite id="xr10e"><a href="#zotero%7C1912747%2FD7ZAZP52">(Candy, 1991)</a></cite>, and<cite id="xs3zn"><a href="#zotero%7C1912747%2FSRHE99CT">(Cropley &#38; Knapper, 2021)</a></cite> have described as self-directed learning as a catalyst for extended learning, both during the course and beyond.

As such, we argue that our digital history course first and foremost serves as a metaphorical ‘gateway drug’ for students towards a development of coding skills and digital history competences, to be further pursued via other channels once the course has been completed. By familiarizing students with several methods in digital history, demonstrating their potential, and giving students some level of hands-on experience in developing a (start of) a fitting code base when adopting them, we aim to persuade the students into deepening their expertise in these methods later on. This, however, can only be achieved when the effort is supported by instruction that fundamentally motivates students to take these further steps. We further discuss our instruction in this regard below.

### Meta-phoria: on opening ‘black boxes’

The ‘wedding seating problem’ has the advantage of offering a simple metaphor to clarify that it can be useful and even necessary to create a ‘workable approximate solution’, rather than a ‘perfect’ or ‘correct’ solution. We then use this example to further discuss the inner workings of (LDA-based) topic modelling, in that the latter also revolves around finding a ‘workable approximate’ distribution of keywords in a given collection of documents to a predefined number of topics (e.g., <cite id="rnwnc"><a href="#zotero%7C1912747%2FIYWZH8W7">(Abdelrazek et al., 2023)</a></cite>Abdelrazek et al., 2023; <cite id="jlyq9"><a href="#zotero%7C1912747%2F3RW5IVDA">(Blei, 2012)</a></cite>Blei, 2012). It has proven easier to explain to students who already understand the wedding seating problem what topic modelling is, how it improves over several iterations and why the results of the model will be slightly different for every student, but nevertheless coherent.

A key issue in digital history education is that it is often challenging for history students to grasp the computational logic underpinning digital methods at first, as this logic is usually new to them. Here, we can connect to our experiences teaching courses on statistics for historians as well, where we have observed similar problems. Students entering undergraduate programs in history, in the Netherlands but also elsewhere, tend to not have any grounding in science, technology, engineering and mathematics (STEM), which creates additional obstacles when trying to integrate mathematical and computational methods-instruction in the program. An element aggravating this is that the available instructional resources on digital history often are still too technical in their design for students starting out with digital history and coding. This creates the impression of an insurmountable inaccessibility, and in turn motivates students to approach the digital methods under discussion as ‘black boxes’, that is, complex instruments whose underlying architectures cannot be understood and should therefore instead be used ‘as is’. As explained above, this runs counter to our educational aims.

In our course, we have increasingly come to address this problem by adopting the following principle: when discussing the subject matter, we move ‘from language to numbers’ rather than the other way around. For example, when introducing a research method, we first discuss it in a metaphorical manner: through the use of analogical reasoning, we, in the words of Black, “bring two separate domains into cognitive and emotional relation by using language directly appropriate to the one as a lens for seeing the other (...) to see a new subject in a new way” (Cited in: (<cite id="ygw99"><a href="#zotero%7C1912747%2F7NVL3K2D">Mahood, 1987)</a></cite>. This, in our experience, has tended to make the material more accessible for students, which serves as a stepping stone to further delve into the complexities of topics at hand.

The session we organize on topic modelling can serve as an illustration. Here, we observed that students struggle with the concept of a probabilistic model, a key principle underlying approaches to topic modelling such as LDA. Initially, we left this computational black box intact, explaining instead how (LDA-based) topic modelling, as well as other probabilistic approaches to text mining, can be used for historical research. However, in practice, this mostly served to further confuse and annoy our students. Therefore, we have come to explain the need for probabilistic modelling in complex tasks in metaphorical terms first, by using the ‘wedding seating problem’, a well-known problem in mathematics and computational science, as an analog and more accessible example outside of text mining, as the entry point for our discussion.

The ‘wedding seating problem’ is, in essence, an optimization problem in light of combinatorial constraints. The idea is that, when planning the seating plan for a wedding, one must find the best possible way to distribute all guests around the tables at the party, so that they all sit next to someone they like. In computational terms, the problem is known as an ‘NP-hard’ problem, in that solving it becomes infeasible very quickly, in that no known (algorithmic) method can reliably find an optimal solution in a reasonable amount of time when the size of the problem increases (e.g. Welch, 1982). The outcome of our discussion is that the best possible solution for the ‘wedding seating problem’ is therefore not a perfect distribution of people, but one that is instead ‘good enough’, achieved by simply ‘starting somewhere’ when assigning places, and by making subsequent iterative improvements.

The ‘wedding seating problem’ has the advantage of offering a simple metaphor to clarify that it can be useful and even necessary to create a ‘workable approximate solution’, rather than a ‘perfect’ or ‘correct’ solution. We then use this example to further discuss the inner workings of (LDA-based) topic modelling, in that the latter also revolves around finding a ‘workable approximate’ distribution of keywords in a given collection of documents to a predefined number of topics (e.g., Abdelrazek et al., 2023; Blei, 2012). It has proven easier to explain to students who already understand the wedding seating problem what topic modelling is, how it improves over several iterations and why the results of the model will be slightly different for every student, but nevertheless coherent.

<cite id="y330k"><a href="#zotero%7C1912747%2FZR4NLMJL">(Kishenblatt-Gimblett &#38; Shandler, 2012)</a></cite>Based on our prior experiences teaching the course, we argue that metaphorical instruction can serve as a powerful tool to bridge initial knowledge gaps in mathematical and computational understanding for students who are less versed in STEM-fields. In evaluations for our course, students indicate that this approach actively helps them to better understand the materials at hand. We therefore recommend this approach to anyone planning to develop instruction on digital history methods.

### Take my stress away: on easing anxiety through instruction and assessment

A key element we have come to observe while teaching our digital history course, is that stress and anxiety can work as major barriers for students to develop an understanding of the computational logic and code base that underpin the digital history methods we teach. As most students who enter our classroom do not have any prior understanding of computational methods and coding, as highlighted above, many students are initially deeply worried that they will not be able to successfully complete the course. Since a primary goal for us is not only to familiarize our students with digital history methods but also motivate them and enable them to further expand their skillset beyond our course, we have therefore come to view addressing this stress and anxiety as primary objectives while teaching. We try to do so in a number of ways.

First, we discuss the matter explicitly with our students throughout the course, to actively engage in expectation management. For example, at the start of the course, we survey our students by asking them how confident they feel in successfully completing the course. We do so by asking them to what extent, on a scale from 1 to 10, they identify with the dog in Figure 1 (a modified version of the well-known ‘doge’ meme-template). Based on this survey question, we engage in an open conversation about the goals and expectations of the course outlined above. Throughout the course, we keep track of the confidence level of our students informally, especially when they further experiment with the course materials, and at the end, we ask whether the students still identify as strongly with the dog in Figure 1 as was the case at the start. This approach, steeped in (admittedly somewhat dated) internet humour, in our experience serves as an effective step in addressing stress and anxiety, while lowering the threshold for students to express their concerns and ask further questions. Moreover, students with low confidence in computing notice that they belong to the majority, which seems to have a positive impact on stress by itself.

Second, and more foundationally, we have come to organize the assessments for the course in such a manner that the focus increasingly lies on individualized trajectories, peer-to-peer learning and formative, rather than summative, evaluation (for the distinction between formative and summative assessment, see originally for example: Scriven, 1967). This serves as a major tool to reduce stress and anxiety among students, as they no longer have to worry about an exam or other formative assessments, where failure would be more definitive. We have increasingly geared our student evaluation towards the question when and where digital methods could be useful for them once they have reached a certain level of proficiency, rather than aiming for a required minimum level of mastery of programming or other computational skills that they need to demonstrate.

As stated above, a major component of the evaluation matrix for the course from the beginning has been the blog-assignment structured around the ‘share your learning’-curve principle, which takes a student’s existing knowledge and skillset as the starting point for learning. This motivates students to further delve into the theories, methodologies and other topics discussed during the course that are of specific interest to them and allows them to do this at their own pace. This, in our experience, has not only resulted in higher levels of intrinsic motivation, but also in reduced levels of stress and anxiety, and as a result, higher levels of self-confidence in learning. In addition, as students are asked to provide feedback on each other’s work as part of the assignment, the chosen approach promotes collaborative learning and mutual support, which also helps to reduce stress and anxiety (see also further below).

We have come to view the second assignment, the research paper, in a similar way. Whereas at first, this assignment was primarily a summative test, in that it served to demonstrate the level of mastery of a chosen digital history method, and its underlying code base, by a student, our choice to introduce it to the students as a ‘feasibility study’ has transformed it into a more formative-oriented assignment. In practice, this means that students hand in both a short paper outlining the research idea, the computational approach they have tried, and their evaluation of their own ideas, and an (R-)script showing how they worked with their data and reached their conclusions. In its current set-up, we still assess whether the students understand and can properly apply the digital method of their choice. In addition, as part of the assignment, we emphasize that students need to develop the scripts for their papers in such a way that they contain extensive explanatory comments, as this allows us to properly assess the operationalized implementation of the project. Apart from this, we are relatively lenient in our assessment of our students’ coding skills, as we use the assignment to primarily provide additional feedback on how to proceed when students are interested to continue using the chosen method.

Our prior experience teaching the digital history course in the Master programs at Erasmus University Rotterdam has shown us that the move away from summative assessment outlined above, has significantly reduced student stress and anxiety, while improving the level of motivation. Moreover, we have found that students become less risk averse when designing projects that are allowed to fail, which in practice leads to better and more adventurous scripts, rather than to a decline in coding. We therefore warmly recommend this approach, especially when the course is aimed at students with an entry-level knowledge of digital history and coding.

### Take my stress away: on easing anxiety through instruction and assessment

Since we started teaching our course on digital history, we have not only observed that students showcase significant stress and anxiety when anticipating summative assessments such as an exam. We have equally come to observe that students quickly get discouraged when making a mistake in class, especially when engaging in coding with R and RStudio. When students encounter a warning or error message when running pre-prepared scripts or when tailoring the latter to their personal research interests in RStudio, students quickly think they have run into an incomprehensible obstacle, and that they cannot proceed.

We can further contextualize this by discussing how a lot of academic instruction today has come to be set up more generally, in the Netherlands but also elsewhere. In our experience, a lot of contemporary academic instruction is rather rigid and closed-ended in its design, in that it does not encourage experimentation and failure. To put it differently, students are used to a rigid trajectory in which only a strictly prescribed end result is counted as ‘good’. In a political and economic environment where academic institutions have, to a significant extent, been reduced to ‘certificate factories’ that should first and foremost pump out labour market-ready graduates, and where students are confronted with significant financial repercussions when not completing their studies in time, students have all too often internalized risk aversion. Consequently, they often do not view mistakes as learning moments that are necessary for growth, but rather as obstacles to the completion of a course, and therefore as something that should be avoided at all costs. For a course where hands-on coding serves as one of the core pillars, this internalized sense of risk aversion is un- or even counterproductive, as it actively works to the detriment of the instruction.

Given this observation, we have come to dedicate a lot of time during our course making students feel comfortable with the general notion that, when programming with R/RStudio, errors are not failures on their part, but rather challenges to be solved through trial and error. This, in our experience, serves as a necessary step when introducing students with entry-level knowledge and experience to coding. Designing motivating assessments is one aspect of this effort. We try to introduce our students to a different mindset and workflow during the course in several other ways as well.

First, when setting up our hands-on coding sessions, we explicitly discuss the different mindset and workflow needed for coding by introducing two broader metaphorical reference points: the game Dark Souls and what we have come to call the ‘frustration to happiness’-graph (and the ‘level of frustration’-parameter) for coding.

Dark Souls is an action-oriented fantasy role-playing game (RPG) created by Japanese game developer FromSoftware in 2011. The game lets players take on the role of a cursed undead character in the fictional kingdom of Lordran, as they set out on a quest to discover their initially unknown fate. The game, as well as the broader ‘Soulsborne’-genre it spawned in the years since, has primarily become known for their high difficulty. Enemy characters in a Dark Souls-like game can be extremely punishing towards the player, resulting in frequent virtual deaths and a loss of a significant amount of progress throughout the game every time an enemy defeats the player. The game has been analysed from multiple perspectives over the past years (e.g., Dooghan, 2025). Some authors have even identified Dark Souls as a game that can help cope with depression, as it motivates players to overcome harsh challenges through endurance (e.g., Gault, 2016). What matters most in our context is that the gameplay in Dark Souls focusses on learning by failure and that eventually overcoming seemingly endless failure is the very point of the game and the thing that players supposedly enjoy.

We have come to adopt the game as a reference point when introducing students to coding with R/RStudio, for two reasons. On the one hand, Dark Souls serves as a humorously exaggerated example when discussing the importance of trial-and-error when coding. This in turn helps lowering the threshold for students to adopt such a mindset when engaging with the course materials. Figure 2 can serve as an illustration of this; whenever a player ‘dies’ in Dark Souls, the words ‘You Died’ gradually expand on the screen, serving as a virtual ‘taunt’ for players to either try again or give up.

On the other hand, we introduce Dark Souls as a reference because the game, and many others alongside it, fittingly incorporate the approach we want to promote among our students. Here, we draw from the body of scholarship surrounding the uses of games for learning that has emerged since the turn of the century. As highlighted by Gee (2003), one of the pioneers in this field, games such as Dark Souls excel at productively integrating learning principles in their designs that promote the development of problem-solving skills, critical thinking, and reflective practices – key skills we want our students to develop when engaging within digital history methodologies and coding in particular. What is more, the RStudio IDE incorporates these principles to some extent as well, in that RStudio instantly provides feedback when a line of code cannot be executed. When setting up the coding sessions in our classroom, we instruct students to use RStudio’s console feedback as a starting point for self-directed learning (e.g., <cite id="56bv5"><a href="#zotero%7C1912747%2FHY79A3XX">(Boud, 2012)</a></cite>; <cite id="myjad"><a href="#zotero%7C1912747%2FD7ZAZP52">(Candy, 1991)</a></cite>; <cite id="967l9"><a href="#zotero%7C1912747%2FSRHE99CT">(Cropley &#38; Knapper, 2021)</a></cite>; Knowles, 1975), for example by consulting supporting resources such as the programming community website Stack Overflow.

To support our emphasis on the need for a troubleshooting mindset and workflow when coding, we have also introduced the ‘frustration to happiness’-graph, as a visualization of the former (Figure 3). The graph metaphorically shows how a learning curve for entry-level coding tends to unfold. As students are initially unfamiliar with coding, and with many of the features of the R-programming environment, they will encounter major obstacles and an accompanying growing level of frustration as a result. However, once they have overcome major obstacles encountered earlier, the experienced level of gratification can be much higher, as is the case when playing difficult games such as Dark Souls. In our instruction, we also encourage our students to report on their personal ‘frustration to happiness’-graphs when writing blog posts structured around the ‘share your learning curve’-principle, something students have done repeatedly and productively over the years.

Second, apart from introducing new metaphorical reference points to prime students for error when coding, we have come to integrate an important social component to our instruction as well. As described above, we already added a peer feedback-component to the ‘share your learning curve’-assignment for the course, as students need to comment on the contributions of at least two fellow students every time they post to complete the assignment. Apart from this, we integrate the social dimension in class as well.

When organizing the demonstrations and workshops in class, we ask our students to form groups of approximately five students, who can help each other when encountering problems while coding. When possible, we also ask students with prior coding experience to spread across the classroom, so that they can help less-experienced students. In that way, we create a set-up that actively capitalizes on, what Vygotsky (1965, 1978), first characterized as a learner’s ‘zone of proximal development,’ that is, the potential development, learning and problem-solving a learner can achieve when helped by others, as opposed to what the learner can achieve independently, without further assistance. During our time teaching the course on digital history, we have observed that students make significant gains when working in groups, and that students continue the group work outside the classroom, to make use of peer support when completing the assignments for the course. It is an approach that equally helps to motivate students to ‘dust themselves off and type again’.

In sum, while teaching the digital history course, we have learned that students with entry-level experience with digital history coding are, in need of a mindset and workflow that academic instruction in its current form often does not promote. To introduce the students to this mindset and workflow, we actively use humour to lower the threshold and make the instruction more accessible. To further promote a trial-and-error approach to coding, we also actively structure our teaching around group work, so that students can motivate and help each other, under our supervision.

### Can we play with madness? on IT-infrastructures and how to navigate them

A final element we wish to highlight when discussing some of the pedagogies and didactics underpinning our digital history course at Erasmus University Rotterdam, relates to the infrastructural context in which we operate while teaching. Over the years, we have come to view many of the physical and digital infrastructures that are available to us while teaching as obstructing, rather than supporting our teaching efforts. This, in turn, requires several proactive steps to make teaching on digital history methods and coding more productive and rewarding for students. From discussions with colleagues at other institutions, we have found that our experience is likely typical.

First, concerning the choice of software for our course, we are strong proponents of using open-source options when possible, such as R and its IDE RStudio for coding. These applications are free, permanently in development through community support, and therefore, more democratic, accessible, and viable in the long run, also when students want to continue working with them once they have completed their program. However, what we have noticed over the years while teaching our course is that structural developments at universities (in the Netherlands and beyond) are actively working against this open source-ethos.

A key development in this respect is the increasing hold of tech giants, active in the publishing- and software sectors in particular, on the infrastructures implemented at Dutch universities. This development is visible in many areas of academic life, but has a tangible impact on how we teach digital history as well.

The clearest example of this is Microsoft, which has several exclusive software licensing agreements with Erasmus University Rotterdam. Over the past few years, Microsoft has fully transitioned towards cloud-based services. This has been accompanied by increasingly restrictive procedures concerning access to services provided at the university, making teaching on digital history and coding more difficult. What is worse, Microsoft not only provides (non-free) software licenses, but also sells educational institutions off-the-shelf digital security solutions. These, understandably, focus on limiting the ability of individual users to make changes, install software and do other things that fall outside the relatively restricted set of actions made possible via, for example, Microsoft Office. This has severely hindered our ability to use applications such as R over the years.

A second infrastructural problem relates to the computational power we have at our disposal at Erasmus University Rotterdam, an issue all digital history educators are likely to encounter when teaching in (under)graduate history programs. Most digital history research projects ‘in the wild’ incorporate very large datasets, and consume considerable amounts of computing power to process them – computing power that most of the time is not available to us. In addition, executing code, for example to train word embeddings, often requires considerable time. Given that each of our sessions consist of three times forty-five minutes in total, it is not feasible for us to download and process large datasets or undertake projects that takes significant amounts of time in class itself.

In light of these constraints, we have come to adopt a more limited overall approach throughout our course. Almost everything we demonstrate in class are scaled-down ‘dry runs’, where we take a small dataset or a limited selection of source materials to demonstrate how a discussed digital research method works in practice. We then further contextualize the results of our live demonstration by highlighting the importance of scale, in particular when using methods such as sentiment analysis, topic modeling and the training of word embedding models. This has worked well over the years to showcase the potential of various methods, although scale does remain desirable.

In sum, while teaching the digital history course the past five years, we have come to realize that the digital infrastructures available to us are often lacking or even actively working against us when trying to implement digital history education. While we have found ways to circumvent some of the related problems, it remains an ongoing topic of concern.

## Conclusion: Happiness, At Last? Where To Go From Here

In this article, we provided a discussion of several key lessons we learned while teaching a course on digital history in the Master history programs offered at Erasmus University Rotterdam in the Netherlands. We did so because we want to contribute to a more widespread adoption of digital history methods by historians – something we think can best be achieved by structurally embedding them in how we train the upcoming generation. Overall, we argue that a reorientation of how (under)graduate history programs are currently set up is a necessity to achieve this goal in the long run. However, as this is not yet the case, we wanted to share what we have learned so far by developing digital history instruction in practice.

Throughout this article, we highlighted how we view our course as both a way to cultivate computational literacy, and as a metaphorical ‘gateway drug’ for students to become invested in digital history and coding. Next, we discussed how we adopt metaphorical instruction to render the discussed digital methods in class intelligible for students who often do not have a background in STEM. We also explained that we take significant steps to lower the stress experienced by students, so they can focus more on engaging and experimenting with the course materials and make them their own. We discussed how we ‘prime students for error’, and finally, we emphasized how university infrastructures sometimes actively hamper setting up a course on digital history methodologies.

Taking the steps above have helped us to make students feel more invested in the course materials, and metaphorically take their first steps on the ‘road to happiness’ towards a better understanding of digital history and coding. At the same time, we see lingering problems that require further addressing. In the final section of the paper, we therefore briefly want to touch on these problems. We identify two broader issues: (1) the integration of digital history education in the program as a whole; (2) the issue of our personal ‘learning curves’ as teachers, and the tension between implementing state-of-the-art methodologies in class versus introducing basic models.

First, while we have taught our course on digital history for several years now, a wider adoption of digital history methodologies throughout (under)graduate history programs as a whole remains a challenge. For example, most of the courses in the Master history programs offered at our university, including interactive workshop sessions organized to support students while writing their Master theses, still tend to not integrate the methods we teach and instead emphasize other approaches. We argue that this issue can be best addressed by rethinking methodological instruction throughout programs in their entirety, to make methodological instruction, including on digital methods, a core component of the full tracks. This, however, is very much still a work in progress, and much work still needs to be done in this area.

Second, there is the issue of our personal learning trajectories as teachers over the years, and how these increasingly diverge from our students with entry-level experience with digital history and coding. Given that we have become more proficient in the use of certain methods, a tendency has been to put more elaborate set-ups in our course. This has worked in some cases, for example while teaching on the use of sentiment analysis, but in others, it has added additional barriers for students. This remains a key point of consideration. As the field is developing at a neck-braking pace, a tendency might become to include as many innovative methods in the course as possible. However, this can significantly increase the complexity of the course and make the learning experience for our students much harder. Our inclination so far has been to stick to more basic approaches.

The points above are but some of the challenges we still see on our ‘road to happiness’ towards the ‘brave new world’ of integrated digital historical research competence hinted at by Zaagsma <cite id="qxhsr"><a href="#zotero%7C1912747%2FCP55TAX5">(2013)</a></cite>) over a decade ago. However, via our discussion here, we hope to have at least helped pave part of the way there.
